```python
file_num = 360
scour = np.zeros((file_num+1,3))
for ifile in range(file_num+1):
    file_str = "sed_"+str(ifile)+".csv"
    ws = pd.read_csv(file_str, header=1, sep=',', skiprows=None)
    ws.columns = ['Terrain Elevation', 'Points:0', 'Points:1', 'Points:2']
    scour[ifile,1] = np.min(ws["Terrain Elevation"])
    scour[ifile,2] = np.max(ws["Terrain Elevation"])
    scour[ifile,0] = ifile*5.0
    print("Finished "+file_str)

np.savetxt(r'./scour_hole_deposition.txt', scour)
```
